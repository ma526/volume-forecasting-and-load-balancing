```python
import pandas as pd
import numpy as np

# Load data
forecast_df = pd.read_csv("../outputs/forecast_output_for_powerbi.csv")
df_partners = pd.read_csv("../outputs/clustered_partners_for_allocation.csv")

# Normalize cost and on-time performance
df_partners['normalized_cost'] = (
    df_partners['cost_per_package'] - df_partners['cost_per_package'].min()
) / (df_partners['cost_per_package'].max() - df_partners['cost_per_package'].min())

df_partners['normalized_on_time'] = (
    df_partners['on_time_percent'] - df_partners['on_time_percent'].min()
) / (df_partners['on_time_percent'].max() - df_partners['on_time_percent'].min())

df_partners['score'] = (
    0.3 * df_partners['normalized_cost'] +
    0.7 * (1 - df_partners['normalized_on_time'])
)

# Aggregate demand at date + sku + zip level
demand_df = forecast_df.groupby(['date', 'sku', 'zip_code'])['volume'].sum().reset_index()

# Initialize capacity tracker
results = []
all_dates = demand_df['date'].unique()

for date in all_dates:
    # Reset partner capacity for the day
    partner_cap = df_partners[['partner_id', 'daily_capacity']].copy()
    partner_cap['remaining_capacity'] = partner_cap['daily_capacity']
    demand_slice = demand_df[demand_df['date'] == date]
    print(date)
    for _, row in demand_slice.iterrows():
        sku = row['sku']
        zip_code = row['zip_code']
        demand = row['volume']
        remaining_demand = demand

        # Sort partners by score
        sorted_partners = df_partners.sort_values(by='score').reset_index(drop=True)
        
        for i, partner in sorted_partners.iterrows():
            partner_id = partner['partner_id']
            tier = partner['tier']
            cost = partner['cost_per_package']
            on_time = partner['on_time_percent']
            score = partner['score']

            cap_row = partner_cap[partner_cap['partner_id'] == partner_id]
            available_cap = cap_row['remaining_capacity'].values[0]

            if available_cap <= 0 or remaining_demand <= 0:
                continue

            alloc = min(available_cap, remaining_demand)

            if alloc > 0:
                results.append({
                    'date': date,
                    'sku': sku,
                    'zip_code': zip_code,
                    'partner_id': partner_id,
                    'allocated_volume': int(alloc),
                    'tier': tier,
                    'cost_per_package': cost,
                    'on_time_percent': on_time,
                    'score': round(score, 4)
                })

                # Update remaining volumes
                partner_cap.loc[partner_cap['partner_id'] == partner_id, 'remaining_capacity'] -= alloc
                remaining_demand -= alloc

        if remaining_demand > 0:
            print(f"‚ö†Ô∏è Unallocated {remaining_demand} for {sku}-{zip_code} on {date}")

# Save result
allocation_df = pd.DataFrame(results)
allocation_df.to_csv("../outputs/load_balanced_allocation.csv", index=False)
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# Load data
forecast_df = pd.read_csv("../outputs/forecast_output_for_powerbi.csv")
df_partners_orig = pd.read_csv("../outputs/clustered_partners_for_allocation.csv")

# Aggregate daily forecast
daily_demand = forecast_df.groupby('date')['volume'].sum().reset_index()
daily_demand.columns = ['date', 'total_volume']

# Normalize partner features
df_partners_orig['normalized_cost'] = (
    df_partners_orig['cost_per_package'] - df_partners_orig['cost_per_package'].min()
) / (df_partners_orig['cost_per_package'].max() - df_partners_orig['cost_per_package'].min())

df_partners_orig['normalized_on_time'] = (
    df_partners_orig['on_time_percent'] - df_partners_orig['on_time_percent'].min()
) / (df_partners_orig['on_time_percent'].max() - df_partners_orig['on_time_percent'].min())

# Storage for results
analysis_results = []

# Try different weights from 0.0 to 1.0 for cost weight
for w in np.arange(0.0, 1.05, 0.1):
    cost_weight = round(w, 2)
    service_weight = round(1 - w, 2)

    df_partners = df_partners_orig.copy()
    df_partners['score'] = (
        cost_weight * df_partners['normalized_cost'] +
        service_weight * (1 - df_partners['normalized_on_time'])
    )

    total_cost = 0
    total_volume = 0
    weighted_service = 0

    for _, row in daily_demand.iterrows():
        date = row['date']
        demand = row['total_volume']
        remaining_volume = demand

        sorted_partners = df_partners.sort_values(by='score').reset_index(drop=True)

        for _, partner in sorted_partners.iterrows():
            if remaining_volume <= 0:
                break

            alloc = min(remaining_volume, partner['daily_capacity'])

            if alloc > 0:
                total_volume += alloc
                total_cost += alloc * partner['cost_per_package']
                weighted_service += alloc * partner['on_time_percent']
                remaining_volume -= alloc

    avg_cost_per_package = total_cost / total_volume if total_volume else 0
    avg_service_level = weighted_service / total_volume if total_volume else 0

    analysis_results.append({
        'cost_weight': cost_weight,
        'service_weight': service_weight,
        'avg_cost_per_package': avg_cost_per_package,
        'avg_service_level': avg_service_level
    })

# Create DataFrame and Plot
df_analysis = pd.DataFrame(analysis_results)

# Plot
fig, ax1 = plt.subplots(figsize=(10, 6))

ax1.plot(df_analysis['cost_weight'], df_analysis['avg_cost_per_package'], 'r-o', label='Avg Cost per Package')
ax1.set_xlabel('Cost Weight in Scoring Formula')
ax1.set_ylabel('Avg Cost per Package', color='r')
ax1.tick_params(axis='y', labelcolor='r')

ax2 = ax1.twinx()
ax2.plot(df_analysis['cost_weight'], df_analysis['avg_service_level'], 'b-s', label='Avg Service Level')
ax2.set_ylabel('Avg Service Level (%)', color='b')
ax2.tick_params(axis='y', labelcolor='b')

plt.title('Trade-off Between Cost and Service Level')
fig.tight_layout()
plt.grid(True)
plt.show()
```
